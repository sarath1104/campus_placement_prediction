DATA PREPROCESSING

```python
m=data['Branch'].unique()
map={}
for i in range(len(m)):
    map[m[i]]=i
for x in range(len(data['Branch'])):
    data['Branch'][x]=map[data['Branch'][x]]
```

```python
map
```

```python
m=data['College'].unique()
map={}
for i in range(len(m)):
    map[m[i]]=i
for x in range(len(data['College'])):
    data['College'][x]=map[data['College'][x]]
```

```python
score=[]
for x in data['skill_set']:
    m=list(x.split(','))
    score.append(len(m))
```

```python
data['score']=score
```

```python
convert_dict = {'Branch': int,'College': int
}
```

```python
data=data.astype(convert_dict)
data
```

```python
x=data[['Branch','College','Percentage/CGPA','internship','backlog','score']]
y=data['placed']
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=1)
```

KNN-CLASSIFICATION

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

TRAIN ACCURACY

```python
model.score(X_train, y_train)
```

TEST ACCURACY

```python
model.score(X_test, y_test)
```

RANDOM FOREST CLASSIFICATION

```python
from sklearn.ensemble import RandomForestClassifier as rf
```

```python
model=rf()
model.fit(X_train, y_train)
```

TRAIN ACCURACY

```python
model.score(X_train, y_train)
```

TEST ACCURACY

```python
model.score(X_test, y_test)
```
